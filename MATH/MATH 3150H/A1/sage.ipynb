```python
# Q4 a

q = 5
p = 1
r = 1

n, x = var('n x')

f_ext = piecewise([
    ((-2 * p, -p), -e ^ (-q * x)),
    ((-p, 0), r),
    ((0, p), e ^ (-q * x)),
    ((p, 2 * p), -r),
    ((2 * p, 3 * p), -e ^ (-q * x)),
])
# plot function s dashed line to see convergence
func_sin = plot(f_ext, (x, -p, p), color='black', linestyle='dashed')

N = [1,5,8,500] #list of N values to plot over
approx_sin = [] #initialize array of plots
L = p #Define length

a_0 = (1/(2*L)) * (integral(r, x, -L, 0)+integral(e^(-q*x), x, 0, L))
a(n) = (1/L) * (integral(r*cos(n*pi*x/L), x, -L, 0)+integral((e^(-q*x))*cos(n*pi*x/L), x, 0, L))
b(n) = (1/L) * (integral(r*sin(n*pi*x/L), x, -L, 0)+integral((e^(-q*x))*sin(n*pi*x/L), x, 0, L))

for i in range (4):
    # Sin only doesn't converge here???
    # g(x) = sum((b(n)*sin(n*pi*x/L)) for n in (1..(N[i]+1)))
    g(x) = a_0 + sum((a(n)*cos(n*pi*x/L) + b(n)*sin(n*pi*x/L)) for n in (1..(N[i]+1)))
    approx_sin += [plot(g, (x, -2*p, 2*p))] #compute and load up plots into array


for i in range (4):
    print("{}th partial sum of Fourier Sine series".format(N[i]))
    show(approx_sin[i] + func_sin) #print out plots with labels
```

```python
# Q5 a
clear_vars()
q = 5
p = 9
r = 22
n, x = var('n x')

f = ((p*x/r)^2)+q
# plot function s dashed line to see convergence
f_ext = piecewise([((-r, 0), -f), ((0,r), f)])
func_sin = plot(f_ext, (x, -r, r), color='black', linestyle='dashed')
N = [5, 10, 50, 100] #list of N values to plot over
approx_sin = [] #initialize array of plots
L = r #Define length

b(n) = (2/L) * (integral((f*sin(n*pi*x/L)), x, 0, L))
for i in range (len(N)):
    g(x) = sum((b(n)*sin(n*pi*x/L)) for n in (1..(N[i]+1)))
    approx_sin += [plot(g, (x, -r, r))] #compute and load up plots into array


for i in range (len(N)):
    print("{}th partial sum of Fourier Sine series".format(N[i]))
    gibbs_upper = plot(f(x=r) *1.18, (x, -r, r), color="red") 
    gibbs_lower = plot(-f(x=r) *1.18, (x, -r, r), color="red")
    show(approx_sin[i] + func_sin + gibbs_upper + gibbs_lower) #print out plots with labels
```
