```python
x, n = var('x n')
assume(n, "integer")

fn = (x^2-2*x)*sin((n*pi*x)/2)

show(integral(fn, x, 0, 2).full_simplify())
```

```python
clear_vars()
x, n, p = var('x n p')
assume(n, "integer")

expr = (x^2 + p*x +2) * sin(pi * n * x)

show((2*integral(expr, x, 0, 1)).full_simplify())
```
