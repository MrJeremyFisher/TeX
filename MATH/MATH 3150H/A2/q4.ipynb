```python
# a
clear_vars()
p = 9
q = 5
r = 22

x, n, t = var("x n t")
h = q
kappa = p
L = q*p
k = p/(q+p)
T1 = 0 # Assumed because this is in the sample code
T0 = 0
f = 10*x*(L-x)*sin(pi*x/L) 
bn(s) = integral(f*sin(s*x), x, 0, L)/integral(sin(s*x)^2, x, 0, L)

w = 0
for n in [1..20]:
    wn=w
    EQ=tan(x*L)+kappa*x/h
    lambn=EQ.find_root((2*n-1)*pi/(2*L),n*pi/L)
    bnn=N(bn(lambn))
    w=wn+bnn*sin(lambn*x)*e^(-lambn^2*k*t)
    show(rf"lambda_{n}: {lambn}")
```

```python
# b
v=T0+x*h*(T1-T0)/(kappa+h*L)
u(x,t)=w+v
```

```python
# c
EQ1 = u(L,t)
plot(EQ1,t,0,120)
```

```python
# d
maxtime, maxtemp  = find_local_maximum(EQ1, 60, 120) # Guessed range based on graph

show(rf"max temp occurs at {maxtime}s and is {maxtemp} units")
```

```python
# e
fun = EQ1 - 25 # will cross at T=25 units
plot(fun, t, 0, 120)
tft = fun.find_root(0,10)
show(rf"25 deg occurs at {tft}s")
```
